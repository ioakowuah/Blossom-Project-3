```python
#Dropping the EmployeeID feature
dataset=dataset.drop('EmployeeNo',axis=1)
dataset
```

```python
#Creating feature of True Promotion
#This is based on Last performance score with less than 7 not getting promotion
def age_groups(series):
    if series <= 7:
        return 0
    elif series > 7:
        return 1

dataset['True_Promotion'] = dataset['Last_performance_score'].apply(age_groups)
```

```python
#Feature engineering of Year of birth
#With base year being 2019, we subtract from indicated year_of_birth in dataset
dataset['Year_of_birth'] = 2019 - dataset['Year_of_birth']
dataset['Year_of_birth']
```

```python
#Feature engineering of Year of recuitment
#With base year being 2019, we subtract from indicated year_of_recruitment in dataset
dataset['Year_of_recruitment'] = 2019 - dataset['Year_of_recruitment']
dataset['Year_of_recruitment']
```

```python
#Function to detect numerical and categorical Variables and impute with median and mode respectively if there is 
#missing values. Here the feature (Qualification) with missing data will be imputted with mode.
def PP_impute_with_medianormode (df):
    import pandas as pd
    import numpy as np
    num_cols = df._get_numeric_data().columns
    categorical_columns=list(set(df.columns) - set(num_cols))
    #print("categorical_columns are",categorical_columns)
     # Get list of DataFrame column names
    cols = list(df)
     # Loop through columns
    for column in cols:
     # Transfer column to independent series
     col_data = df[column]
     # Look to see if there is any missing numerical data
     
     missing_data = sum(col_data.isna())
     if missing_data > 0:
        if(column in categorical_columns): #For categorical columns replace with mode
     #col_mode=col_data.mode()
            col_mode=col_data.value_counts().index[0]
     #print("Categorical column is '%s' and its mode is '%s'"%(column,col_mode))
            col_data.fillna(col_mode, inplace=True)
     #col_data.replace('NaN',col_mode,inplace=True)
     df[column] = col_data
    else:
     # Get median and replace missing numerical data with median
     col_median = col_data.median()
     #print("Numerical column is '%s' and its median is '%s'"%(column,col_median))
     col_data.fillna(col_median, inplace=True)
     df[column] = col_data
     return df
```

```python
#Detecting numerical features from data
num_cols = dataset._get_numeric_data().columns
num_cols
```

```python
#Detecting categorical features in data
categorical_columns=list(set(dataset.columns) - set(num_cols))
categorical_columns
```

```python
#Storing imputed dataframe as dfimputed
dfimputed = PP_impute_with_medianormode(dataset)
dfimputed.info()
```

```python
#Encoding categorical features with onehotencoding
dfonehotencoded= pd.get_dummies(dfimputed, prefix_sep='_',columns=categorical_columns)
dfonehotencoded.info()
```

```python
#Checking Imbalance representation in target feature
dataset['Promoted_or_Not'].value_counts()
dfonehotencoded['Promoted_or_Not'].value_counts()
```

```python
#Percentage of imbalance data in target feature
print('0 (Non Promoted)', round(dfonehotencoded['Promoted_or_Not'].value_counts()[0]/len(dfonehotencoded) * 100,2), '% of the dataset')
print('1 (Promoted)', round(dfonehotencoded['Promoted_or_Not'].value_counts()[1]/len(dfonehotencoded) * 100,2), '% of the dataset')
```

```python
# Separate input features (X) and target variable (y)
from sklearn.model_selection import train_test_split
y = dfonehotencoded.Promoted_or_Not
X = dfonehotencoded.drop('Promoted_or_Not', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
#Treating imbalance data with resample - upsample minority class
from sklearn.utils import resample
# concatenate our training data back together
X = pd.concat([X_train, y_train], axis=1)

# Separate majority and minority classes
X_majority = X[X.Promoted_or_Not==0]
X_minority = X[X.Promoted_or_Not==1]
 
# Upsample minority class
minority_upsampled = resample(X_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=len(X_majority),    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
dfonehotencoded_upsampled = pd.concat([X_majority, minority_upsampled])

# Display new class counts
dfonehotencoded_upsampled.Promoted_or_Not.value_counts()
```

### Using Up Sample Minority class is the method I decide to use for my imbalance data. I believe using this method will provide more data which is good for modelling

# Question 5 

```python
#Logistics Regression
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score
log_reg=LogisticRegression(solver='liblinear')
log_reg.fit(X_train,y_train)
training_score = cross_val_score(LogisticRegression(solver='liblinear'), X_train, y_train, cv=5)
y_pred_log = log_reg.predict(X_test)

print('---' * 45)
print('Logistic Regression: \n')
print('Recall Score: {:.2f}'.format(recall_score(y_test, y_pred_log)))
print('Precision Score: {:.2f}'.format(precision_score(y_test, y_pred_log)))
print('F1 Score: {:.2f}'.format(f1_score(y_test, y_pred_log)))
print('Accuracy Score: {:.2f}'.format(accuracy_score(y_test, y_pred_log)))
print("Cross Validation Accuracy ", round(training_score.mean(), 2) )
print('---' * 45)
```

```python
#Barplot of Feature Importance of SVM Classifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC
def plot_coefficients(classifier, feature_names, top_features=20):
 coef = classifier.coef_.ravel()
 top_positive_coefficients = np.argsort(coef)[-top_features:]
 top_negative_coefficients = np.argsort(coef)[:top_features]
 top_coefficients = np.hstack([top_negative_coefficients, top_positive_coefficients])
 # create plot
 plt.figure(figsize=(15, 5))
 colors = ['red' if c < 0 else 'blue' for c in coef[top_coefficients]]
 plt.bar(np.arange(2 * top_features), coef[top_coefficients], color=colors)
 feature_names = np.array(feature_names)
 plt.xticks(np.arange(1, 1 + 2 * top_features), feature_names[top_coefficients], rotation=60, ha='right')
 plt.show()
cv = CountVectorizer()
cv.fit(dfonehotencoded_upsampled)

svm = LinearSVC()
svm.fit(X_train,y_train)
plot_coefficients(svm, cv.get_feature_names())
```
